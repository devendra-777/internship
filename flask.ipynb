```python
from flask import Flask, request, jsonify
import random, datetime

app = Flask(__name__)

FUN_FACTS = [
    "Python was named after Monty Python, not the snake.",
    "Flask is a lightweight web framework—perfect for quick apps.",
    "Uppercase adds emphasis—great for usernames and titles.",
    "Data Science blends statistics, programming, and storytelling.",
    "APIs let apps talk to each other—your browser is a client too."
]

def stylize(text: str) -> str:
    return " ".join(char.upper() if i % 2 == 0 else char.lower() for i, char in enumerate(text))

def vowel_count(text: str) -> int:
    return sum(1 for c in text.lower() if c in "aeiou")

@app.route("/")
def home():
    name = request.args.get("name", "Guest")
    upper_name = name.upper()
    reversed_name = name[::-1]
    char_count = len(name)
    vowels = vowel_count(name)
    fact = random.choice(FUN_FACTS)
    now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    return f"<h1>Hello, {upper_name}!</h1><p>Reversed: {reversed_name}</p><p>Chars: {char_count}</p><p>Vowels: {vowels}</p><p>{fact}</p><p>Server time: {now}</p>"

@app.route("/api/transform")
def api_transform():
    name = request.args.get("name", "Guest")
    return jsonify({
        "original": name,
        "uppercase": name.upper(),
        "reversed": name[::-1],
        "stylized": stylize(name),
        "length": len(name),
        "vowel_count": vowel_count(name),
        "fun_fact": random.choice(FUN_FACTS)
    })

@app.route("/shout/<name>")
def shout(name):
    return name.upper()
```

```python
app.run(port=5000)
```
